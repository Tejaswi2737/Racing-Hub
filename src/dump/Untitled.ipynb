```python
import pandas as pd 
df=pd.read_json('today_json.json')
```

```python
df_exp=df['pool_fh'].str.split("_",expand=True)
df_exp
```

```python
df_exp.columns=['Date','category','Type','Place','Slot','Slot_Char']
```

```python
df_exp.shape
```

```python
df_all=pd.concat([df,df_exp],axis=1,sort="False")
df_all
```

```python
df_all=df_all[['pool_status','Date','category','Type','Place','Slot']]
```

```python
df_greyhound=df_all.loc[df_all['Type']=="greyhound"].reset_index(inplace=False)
df_thoroughbred=df_all.loc[df_all['Type']=="thoroughbred"].reset_index(inplace=False)
```

```python
df_thoroughbred
```

```python
grey_slots=df_greyhound['Slot'].unique()
thorough_slots=df_thoroughbred['Slot'].unique()
grey_place=df_greyhound['Place'].unique()
thorough_place=df_thoroughbred['Place'].unique()
```

```python
df_table_grey=pd.DataFrame(index=grey_place,columns=grey_slots)
df_table_thorough=pd.DataFrame(index=thorough_place,columns=thorough_slots)
```

```python
df_table_grey.columns=range(1,len(grey_slots)+1)
df_table_thorough.columns=range(1,len(thorough_slots)+1)
```
